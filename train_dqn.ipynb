```python
from train_dqn import train_dqn
import os
import torch
```

```python
instances = ["la01", "la06", "la11", "la21", "la31", "la36", "orb01", "swv01", "swv06", "swv11", "yn1"]
```

```python
num_episodes = 1000
memory_size = 100000
batch_size = 64
target_update = 100
noisy = False
plotting_inteval = 10
schedule_cycle = 8
```

```python
for instance in instances:
    plot_file = f"plots_training_dynamic/{instance}/num_episodes={num_episodes}_memory_size={memory_size}_target_update={target_update}_noisy={noisy}_cycle={schedule_cycle}.png"
    policy, t = train_dqn(instance, num_episodes=num_episodes, memory_size=memory_size, batch_size=batch_size, 
                            target_update=target_update, noisy=noisy, plotting_interval=plotting_inteval, save_plot=plot_file, schedule_cycle=schedule_cycle)
    policy_file = f"policies/dqn_mlp/{instance}_num_episodes={num_episodes}_memory_size={memory_size}_target_update={target_update}_noisy={noisy}_cycle={schedule_cycle}.pth"
    torch.save(policy.state_dict(), policy_file)
    print(instance, " done.")
```

```python

plot_file = f"plots_training_dynamic/{instance}/num_episodes={num_episodes}_memory_size={memory_size}_target_update={target_update}_noisy={noisy}_cycle={schedule_cycle}.png"
policy, t = train_dqn(instance, num_episodes=num_episodes, memory_size=memory_size, batch_size=batch_size, 
                        target_update=target_update, noisy=noisy, plotting_interval=plotting_inteval, save_plot=plot_file, schedule_cycle=schedule_cycle)
policy_file = f"policies/dqn_mlp/{instance}_num_episodes={num_episodes}_memory_size={memory_size}_target_update={target_update}_noisy={noisy}_cycle={schedule_cycle}.pth"
torch.save(policy.state_dict(), policy_file)
```
