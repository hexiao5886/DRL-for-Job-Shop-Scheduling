```python
import os
import pandas as pd
```

```python
dirs = ["p0.5cv0.2_gin_(20, 20)_10000_updates", "p0.5cv0.2_gin_(30, 20)_10000_updates", \
        "p0.5cv0.2_gin_(50, 10)_2000_updates", "p0.5cv0.2_gin_(50, 10)_500_updates"]
dirs = ["p0.5cv0.2_gin_(20, 20)_10000_updates", "p0.5cv0.2_gin_(30, 20)_10000_updates"]
names = ["(20, 20)_10000_updates", "(30, 20)_10000_updates",\
         "(50, 10)_2000_updates", "(50, 10)_500_updates"]
```

```python
df2 = pd.DataFrame()
for i, directory in enumerate(dirs):
    file = directory + "/data.csv"
    df = pd.read_csv(file)
    makespan = df["policy_on_original"]
    if i == 0:
        instances = df["instance"]
    df2[names[i]] = makespan
    
```

```python
df2.index = range(len(instances))
```

```python
ax = df2.plot(xticks=df.index)
ax.set_xticklabels(instances)
```
